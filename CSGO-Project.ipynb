**CSGO PROJECT**

**COUNTER STRIKE(CS)** is a popular series of tactical first person shooter(FPS) vedio games that have been enjoyedby gamers worldwide for many years.The series is originated from the popular game Half life.

Here is an overview of Counter Strike:

**Gameplay Overview:**

Counter strike is primarily a multiplyare game where two teams . the Counter-Terroists(CTs) and the Terroists(TS) compete against each other.

The objective of each round varies based the game model,but primarly goal includess:

**Counter Terroists:**Prevent the Terroists from achieving their objectives,such as defusing bombs or rescuing the hostages.

**Terroists:**Achieve their objectives which mayinclude planting a bomb at a designate site or holding the hostages.Rounds are relatively short,lastinge few minutes and players have one life per round.When a player is eliminated they must have to wait until the next round to repawn.

Key Fetures:

Weapons:Players can purchase and use a wide variety of firearms,equipmenst.The choice of weapon is an essential startegic element in the game.

Economy:Player earns in game money based on theirs performance in the previous rounds.Money is used to buy wepons or equipments for the next round.

Maps:Counter srike features a range of maps,each with its own layout and objectives.DustII,inferno,Mirage etc.

Teamwork:Successful gamplay of CS relies on teamwork,communication.Players often coordinates their actions with their teammates to achieve objectives.

Popular Games Mode:

Bomb defusal(de_):In this mode,Terroists attempt to plant a bomb at one of the designated bomb sites,while Counter terroists aim to prevent the bomb from being planted or defused if it is planted.

Hostage defuse(cs_):In hostage refuse mode,Counter terroists must rescue hostages held by the terroists.

Arms Races:A fast paced mode where players cycle through a series of weapons,aiming to be the first to get kill with each weapon.

Deathmatch:A mode where playes respawn quickly and aim to get as many kills as possible within set limit of time.

### Building the model with LDA

**Lets proceed with LDA for feature selection but not extraction as we are having only 2 classes in the target variable.**

LDA tries to reduce the number of features by picking the ones that are most useful for distingusihing between the groups.

For example,if you have 10 features,LDA tells you that only 3 of them are really good at seperating the groups.

**Standarize the data**

**LDA is a sensitive to features scales:**

If one feature has a much larger range(eg.salary in thousands vs age in years),it can dominate the model and distotrt the result.

**Scaling the features ensures that all features contribute equally to the analysis by bringing them ontothe same scale.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
new=pd.DataFrame(x_train)
new
```

```python
#LDA implementation

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis()
lda.fit(x_train,y_train)
```

```python
lda.transform(x_test)  # when you run lda.tranform the lda model tranforms the test data(x_test) into a new space  with reduce dimension.
```

Each features in your dataset contributes differently to seperaing the classes.**lda.coef_ provides the importance of each features for each class.**

These coeficients represent the direction and magnitude by which each feature helps in differentiationg the classes.

```python
lda.coef_
```

**lda.coef_**:This retrives the coefficents(or weights) of the features from the trained LDA model.These coefficents indicate the importance of each features in distingushing between classes.

**np.aps(lda.coef_)**:The absolute value of the coefficents is taken.This remove any negative signs,focusing on the magnitude of the feature importance rather than its direction(positive or negative).

**np.exp(np.aps(lda.coef_))**:The exponantial function(e^x where e is Euler's numbers ,approximately 2.718) is applied to each of these absolute values.This operation tranforms the coefficents exponentially by scaling them.

```python
lda_coefficients=np.exp(np.abs(lda.coef_))
lda_coefficients
```

***Why flattening the array?***

**Simplifying the Sturcture:** After calculating the coefficients,you might want to have all the values in a single list instead of multidimesional structure.This makes easire to acces or perform opearions on the coefficents.

**Easy to use:** Many operations (like indexing,sorting etc) areeasire to perform on a 1D array.for example,if lda coefficients is originaly a 2d array (1,n) or (m,n) flattening transform it into a simple list like structure.

```python
lda_coefficients=lda_coefficients.flatten()
```

```python
lda_coefficients
```
